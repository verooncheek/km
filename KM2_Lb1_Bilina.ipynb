# Этап 3. Определение функции перемещения и нахождение угла запуска тела

Полагаем, что начало прямоугольной декартовой системы координат соответствует уровню земли. Тогда в начальный момент движения тела $t=0$ имеем, что $$s_x(0)=0,     s_y(0)=h_{start}$$
Для построения функции перемещения $(s_x(t),s_y(t))$ воспользуемся вторым законом Ньютона $$m\bf a=F,$$ где $m$— масса тела, ускорение движения тела ${\bf a}=(s"_x(t),s"_y(t))$ и приложенная сила ${\bf F}=m(0,-g)$. В результате имеем два уравнения, что $$s"_x(t)=0,s"_y(t)=-g$$. На основании уравнений с учетом начальных условий можно записать
аналитический вид для функций $ s_x(t),s_y(t)$ $$s_x(t)=s_x(0)+v0_xt=0+v0_xt,$$ $$s_y(t)=s_y(0)+v0_yt-\frac g 2 t^2,$$ где ${\bf v0} = (v0_x,v0_y)$ обозначает неизвестную скорость движения тела в начальный
момент времени $t=0$ .


$\color{red}\text{Напишите}$ подробно, как из второго закона Ньютона и начальных
условий получен аналитический вид функции перемещения$(s_x(t),s_y(t))$ .
Объяснения оформите в тексте документа с лабораторной работой.

Полёт тела, брошенного под углом к горизонту, можно представить как комбинацию двух независимых движений: равномерного горизонтального (по оси $O_x$) и равнопеременного вертикального (по оси $O_y$). В начальный момент времени скорость тела имеет проекции $v0_x$ и $v0_y$ на эти оси.

Движение по оси $O_X$ является равномерным, так как на тело не действуют никакие силы в горизонтальном направлении (пренебрегая сопротивлением воздуха). Следовательно, горизонтальная скорость $v0_x$ остаётся постоянной на протяжении всего полёта. Горизонтальная координата тела в любой момент времени $t$ определяется как $s_x$ = $s_x(0)$ + $v_0t$, где $s_x(0)$ – начальная горизонтальная координата.

Вертикальное движение (по оси $O_y$) происходит под воздействием постоянного ускорения свободного падения $g$, направленного вниз. Поэтому, вертикальная скорость тела изменяется со временем. Вертикальная координата тела в любой момент времени $t$ определяется как $s_y = s_y(0) + v0_yt –\frac{gt^2}{2}$, где $s_y(0)$ – начальная вертикальная координата.
Если принять начальную горизонтальную координату за ноль, а начальную вертикальную за высоту $h_{start}$, то $s_x(0) = 0$ и $s_y(0) = h_{start}$.

$\color{red}\text{Задание* необязательное}$ как изменится вид функций
перемещения, если начальный момент времени будет равен не $t=0$, а $t=t^*$?

Если отсчет времени начинается не с момента броска ($t_0 = 0$), а с некоторого момента времени $t^*$, то уравнения движения остаются прежними, но изменяются начальные координаты $s_x(0)$ и $s_y(0)$. Новые начальные координаты можно вычислить, подставив $t^*$ в уравнения для горизонтального и вертикального движения соответственно: $s_x(0) = v0_x^*$'× и $s_y(0) = h_{start} + v0_yt^* – \frac{gt^2}{2}$.

Из условия задачи имеем, что $s_x(T)=s_{end}$. Из этого соотношения находим
значение для $v0_x$

```python
v0_x = s_end/T
v0_x
```

```python
v0_y = (h_end-h_start+g/2*T**2)/T
v0_y
```

Значение угла запуска тела $\alpha$ определяем из соотношения $tan \alpha=\frac{v_{0y}}{v_{0x}}$

```python
alpha = math.atan(v0_y/v0_x)
alpha, math.degrees(alpha)
```

Результирующая функция перемещения имеет следующий вид

```python
f's_x(t) = {v0_x} t'
```

```python
f's_y(t) = {h_start} + {v0_y:.6} t - {g:.6}/2*t\N{superscript two}'
```

Уравнение прямой, проходящей через две точки: $ (x_1, y_1) $ и $ (x_2, y_2) $:  $$ s_{lin} = mt + b $$
где: **Наклон (угловой коэффициент) $ m $**:  
$$ m = \frac{s[k+1] - s[k]}{t_{\text{liststop}}[k+1] - t_{\text{liststop}}[k]}
$$

- **Свободный член $ b $**:  

$$
b = s[k] - m \cdot t_{\text{liststop}}[k]
$$

```python
m = (s2 - s1) / (t2 - t1)
b = s1 - m * t1
t_fall= -b / m
print(t_fall)
```

Для построения касательной прямой к траектории движения тела в момент времени $t = 0$ воспользуемся уравнением прямой, проходящей через точку $p_{\text{start}}(x_{\text{start}}, y_{\text{start}})$ с угловым коэффициентом $k$:$$y = k(x - x_{\text{start}}) + y_{\text{start}}$$

Угловой коэффициент $k$ равен тангенсу угла наклона касательной к графику функции в точке и равен производной $s'(t)$ в точке $t=0$. У нас есть параметрическое представление траектории $(x(t), y(t))$, значит $k$ можно вычислить как:
$$ k = \frac{dy}{dx} = \frac{y'(t)}{x'(t)}$$

при $t = 0$.

### Уравнение нормальной прямой

Нормальная прямая является перпендикулярной касательной и проходит через ту же точку $p_{\text{start}}$. Для нахождения её углового коэффициента воспользуемся свойством перпендикулярности:  

Если две прямые перпендикулярны, то произведение их угловых коэффициентов равно $-1$:

$$
k_{\text{normal}} = -\frac{1}{k}
$$

Тогда уравнение нормальной прямой имеет вид:

$$
y = k_{\text{normal}} (x - x_{\text{start}}) + y_{\text{start}}
$$

```python
s_x=v0_x*t
s_y=h_start+v0_y*t+g*t**2/2
t=sympy.symbols('t')
sdif_x=sympy.diff(s_x, t)
sdif_y=sympy.diff(s_y, t)
k=sdif_y.subs(t,0)/sdif_x.subs(t,0)
y_kasstart = np.array([s_x_array[0],s_y_array[0]])
y_kasend = np.array([s_x_array[0]+1,k+s_y_array[0]])
y_kas=[y_kasstart+ (y_kasend-y_kasstart)*t for t in [-1, 2]]
y_kas = np.array(y_kas)
y_norstart = np.array([s_x_array[0],s_y_array[0]])
y_norend = np.array([s_x_array[0]+1,-1/k+s_y_array[0]])
y_nor=[y_norstart+ (y_norend-y_norstart)*t for t in [-1, 2]]
y_nor = np.array(y_nor)
```
